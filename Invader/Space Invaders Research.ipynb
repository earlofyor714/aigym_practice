### Basic LSTM Model

```python
vs.plot_trials('sim_default-learning.csv')
```

```python
import tensorflow as tf
import numpy as np
import random
```

```python
g = tf.Graph()
with g.as_default():
    q = tf.zeros([1,6], dtype=tf.float32)
    max_q = tf.argmax(q, 1)
    max_val = q[0, max_q[0]]
    maxes = tf.equal(max_val, q[0])
    # maxes is bool table containing locations of max values.  
```

```python
with tf.Session(graph=g) as sess:
    b = sess.run([q])
    print("q: {}".format(b[0][0]))
    print("max_q: {}".format(sess.run([max_q])))
    print("max_val: {}".format(sess.run([max_val])))
    a = sess.run([maxes])
    print("maxes: {}".format(a))
    c = np.where(a)[1]
    print(c)
    print(random.choice(c))
```
